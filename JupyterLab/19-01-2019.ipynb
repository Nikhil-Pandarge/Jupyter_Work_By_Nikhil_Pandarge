- ## Coditional Comprehesion:

```python
num = [10,20,-10,30,40,50,-10] # declration of list as 'num'.

output = [value ** 2 for value in num if value > 10] #comprehesion.

print(output) # printing output
```

```python
output_cod = [value if value % 2 == 0 else 0 for value in num]
print(output_cod)
```

```python
 # print only number which is dividable by 2 else print None
[num if num % 2 == 0 else None for num in range(10)]
```

```python
[num for num in range(10) if num % 2 == 0  ]
```

```python
dict_exapmle = {num:num for num in range(10)}
print(dict_exapmle)
```
