## - Matplotlib using matplotlib.pyplot subpackage

```python
import matplotlib.pyplot as plt #here we import matplotlib
year= [1950,1960,1990,2010] # this is for x co-ordinate
pop= [2.5,3.6,5.2,6.4] # this is for y co-ordinate
plt.plot(year, pop) #first one for x co-ordinate and pop on y-co-ordinate
plt.show()
```

## Scatter
- using plt.scatter()

```python
plt.scatter(year,pop)
plt.show()
```
