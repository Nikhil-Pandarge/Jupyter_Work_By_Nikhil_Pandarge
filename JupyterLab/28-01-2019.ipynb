```python
weather = pd.read_csv('./pydatadc_2018-tidy-master/data/weather.csv')

weather.head(10)
```

```python
weather_melt = pd.melt(weather,id_vars=['id','year','month','element'],value_name='Counts',var_name='Day')

weather_melt.head()
```

```python
weather_pivot = pd.pivot_table(weather_melt,index=['id','year','month','Day'],columns='element',values='Counts')
```

```python
weather_pivot
```

```python
weather_pivot.reset_index()
```
