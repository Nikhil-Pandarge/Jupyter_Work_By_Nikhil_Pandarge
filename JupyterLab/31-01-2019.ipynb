- ## Importing flat files using NumPy

> - NumPy arrays: standard for storing numerical data
> - Essential for other packages: e.g. scikit-learn
> - loadtxt()
> - genfromtxt()

```python
import numpy as np
file_name_MNIST = 'mnist.txt'
data = np.loadtxt(file_name_MNIST,delimiter=',')
print(data)
```

```python
data_skip_1 = np.loadtxt(file_name_MNIST,delimiter=',',skiprows=1)
print(data_skip_1)
```

```python
data_skip_2 = np.loadtxt(file_name_MNIST,delimiter=',')
print(data_skip_2)
```

```python
data_conversion_string = np.loadtxt(file_name_MNIST,delimiter=',',dtype=str)
print(data_conversion_string)
```
