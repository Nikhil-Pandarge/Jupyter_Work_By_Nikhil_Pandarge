```python
import pandas as pd
import matplotlib.pyplot as plt
file_path_stock = 'C:\\Users\\Family\\all_stocks_5yr.csv'
stock = pd.read_csv(file_path_stock,parse_dates=True,index_col='date')
stock.sorted()
```
